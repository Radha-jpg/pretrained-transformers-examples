```python
# Install once: pip install transformers
from transformers import pipeline

# Sentiment analysis (zero-shot, no fine-tuning needed)
classifier = pipeline("sentiment-analysis")
result = classifier("This breakthrough in open AI is incredibly exciting!")
print(result)  # [{'label': 'POSITIVE', 'score': 0.999}]
```

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
result = classifier("The new open-source tools in 2026 are revolutionizing NLP!")
print(result)
# Example output: [{'label': 'POSITIVE', 'score': 0.9990310668945312}]
```

```python
classifier = pipeline("zero-shot-classification")

sequence = "This movie was visually stunning but the plot fell flat."
labels = ["positive", "negative", "neutral"]

result = classifier(sequence, candidate_labels=labels)
print(result['labels'][0], result['scores'][0])
# Likely: negative 0.8962516784667969
```

```python
qa = pipeline("question-answering")

context = "Hugging Face was founded in 2016 in New York and now hosts millions of models."
question = "When was Hugging Face founded?"

result = qa(question=question, context=context)
print(result['answer'])  # "2016"
```

```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('all-MiniLM-L6-v2')  # Fast & popular in 2026

sentences = [
    "Transformers revolutionized NLP.",
    "BERT is a popular encoder model.",
    "Hugging Face provides easy access to models."
]

embeddings = model.encode(sentences)

# Compute similarity between sentence 0 and others
similarities = util.cos_sim(embeddings[0], embeddings)[0]
print(similarities)
# High score for related sentences
```

```python
# Abstractive text summarization using the facebook/bart-large-cnn model from Hugging Face.
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("facebook/bart-large-cnn")
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/bart-large-cnn")

article = """
Artificial intelligence continues to advance rapidly in 2026. Transformers remain the backbone of NLP, with Hugging Face hosting over 2 million models.
Recent improvements include better long-context handling and efficient fine-tuning methods like PEFT. Researchers are excited about multimodal models combining text, vision, and audio for richer analysis.
"""

inputs = tokenizer(article, max_length=1024, return_tensors="pt", truncation=True)

summary_ids = model.generate(
    inputs["input_ids"],
    num_beams=4,
    min_length=30,
    max_length=130,
    early_stopping=True
)

print(tokenizer.decode(summary_ids[0], skip_special_tokens=True))

#Recent improvements include better long-context handling and efficient fine-tuning methods like PEFT. Researchers are excited about multimodal models combining text, vision, and audio.
```
